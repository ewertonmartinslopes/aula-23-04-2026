```python
lista_vendas = [1000,500,800,1500,2000,2300]
meta = 1200
percentual_bonus = 0.1
for venda in lista_vendas:
  if venda >= meta:
    bonus = percentual_bonus * venda
  else:
    bonus = 0
  print(bonus)
```

```python
#ATIVIDADE 01

vendas_funcionarios = [2800,3200,1500,4000,5000,1000,3500]
meta = 3000
percentual_bonus = 0.05
for venda in vendas_funcionarios:
  if venda >= meta:
    bonus = percentual_bonus * venda
  else:
    bonus = 0
  print('bonus:', bonus)
```

```python
#ATIVIDADE 02

vendas_diarias = [2000,3000,2500,4000,1800,5000,2700]
meta = 2500
vendas_acima_da_meta = 0
for dias in vendas_diarias:
  if dias > meta:
    vendas_acima_da_meta += 1
    print(vendas_acima_da_meta)
```

```python
#ATIVIDADE 03

vendas_vendedores =[1500,2500,1800,3200,4000,1200,2800]
meta = 2000
comissao = 0.08
for vendas in vendas_vendedores:
  if vendas > meta:
    bonus = comissao * vendas
  else:
    bonus = 0
  print('sua comissão sera:', bonus)
```

```python
#ATIVIDADE 04

lista_vendas = [1000,500,800,1500,2000,2300]
meta = 1200
contador = 0
for venda in lista_vendas:
  if venda >= meta:
    contador += 1
  print('atingiram a meta:', contador)
```

```python
#ATIVIDADE 05

lista_vendas = [1000,500,800,1500,2000,2300]
meta = 1200
meta2 = 2000
percentual_bonus = 0.1
bonus_extra = 50
for vendas in lista_vendas:
  if vendas >= meta:
    bonus = percentual_bonus * vendas
    if vendas >= meta2:
      bonus = percentual_bonus * vendas + bonus_extra
  else:
    bonus = 0
  print(f'vendas:{vendas} sua comissão será: R${bonus}')
```

```python
#ATIVIDADE 06

lista_vendas = [1000,500,800,1500,2000,2300]

maior_vendas = lista_vendas[0]

for venda in lista_vendas:
  if venda > maior_vendas:
    maior_vendas = venda
print('a maior venda:', maior_vendas)
```
